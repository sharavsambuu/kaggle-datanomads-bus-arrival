```python
filled_df = pd.read_csv('./out/output.csv')
filled_df = filled_df.rename(columns={"TIMESTAMP": "TIMESTAMP_FILLED"})
filled_df
```

```python
test_df = test_df.join(filled_df.set_index('index'), on='index')
test_df
```

```python
test_df['DIFF'] = test_df['TIMESTAMP_FILLED'].shift(-1)-test_df['TIMESTAMP']
```

```python
test_df[test_df['DIFF']<0]
```

```python
test_df['TIMESTAMP_FIX']=test_df.apply(lambda x: x['TIMESTAMP'] if (x['DIFF']<0) else x['TIMESTAMP_FILLED'], axis=1)
```

```python
test_df[['index', 'TIMESTAMP_FIX']].to_csv("./out/output_fix.csv", index=False)
```
