```python
train_df = pd.read_csv('./data/training.csv')
test_df  = pd.read_csv('./data/test.csv')
```

```python
train_buses = train_df['BUS_ID'].unique().tolist()
```

```python
test_buses = test_df['BUS_ID'].unique().tolist()
```

```python
not_found = []
for bt in test_buses:
    if not bt in train_buses:
        print(bt, "is not found in training")
        not_found.append(bt)
```

```python
total = 0
for bid in not_found:
    r = test_df[test_df['BUS_ID']==bid]
    total = total + len(r)
    print(bid, len(r))

print("impact", total, "of", len(test_df))
```
