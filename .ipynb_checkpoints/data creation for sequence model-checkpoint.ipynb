```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv('./out/cleaned_train.csv')
train_df
```

```python
import random
# 5000-н урттай дарааллаас хамгийн ихдээ 50 урттай дэд цонхнууд үүсгэх туршилт
# мөн тэндээс санамсаргүй sample сонгох
def generate_n_sample_indexes(seq_len=5000, window_len=50, sample_size=10):
    l = [n for n in range(1, seq_len)]
    start_idx = 0

    total_sample_indexes = []
    for i in range(start_idx, seq_len-window_len+1):
        for j in range(i+1, i+window_len+1):
            total_sample_indexes.append((i,j))

    return random.choices(total_sample_indexes, k=sample_size)

print(generate_n_sample_indexes())
```

```python
total_generated_samples = 500000
print("total needed samples", total_generated_samples)
print("route count", len(train_df['BUSROUTE_ID'].unique()))
samples_per_route       = int(total_generated_samples/len(train_df['BUSROUTE_ID'].unique()))
print("sample needed per route", samples_per_route)
sample_buses_per_route  = 20
samples_per_bus         = int(samples_per_route/sample_buses_per_route)
print("sample needed per bus", samples_per_bus)
bus_history_treshold    = 500
```

```python
dataset = []

# Бүх чиглэлийг хамрах
for idx, route_id in enumerate(train_df['BUSROUTE_ID'].unique()):
    print("proccessing at route", route_id)
    # энэ чиглэлээс санамсаргүй автобус сонгох гэхдээ тодорхой тооноос дээш бичлэгтэй
    filtered_df = train_df[(train_df['BUSROUTE_ID']==route_id)].groupby("BUS_ID").filter(lambda x: len(x)>bus_history_treshold)
    # шүүлт хийгдсэн бичлэгнээс санамсаргүй хэдэн автобус сонгох
    filtered_buses = filtered_df['BUS_ID'].unique()
    if len(filtered_buses)==0:
        continue
    if len(filtered_buses)<sample_buses_per_route:
        bus_sample_count = len(filtered_buses)
    else:
        bus_sample_count = sample_buses_per_route
    selected_buses = random.choices(filtered_df['BUS_ID'].unique(), k=sample_buses_per_route)
    selected_buses = np.unique(selected_buses).tolist()
    
    # сонгогдсон автобуснуудаас тодорхой тооны явсан түүх үүсгэх
    for bus_id in selected_buses:
        bus_history     = filtered_df[filtered_df['BUS_ID']==bus_id]
        bus_history_len = len(bus_history)
        sample_indexes  = generate_n_sample_indexes(seq_len=bus_history_len, window_len=50, sample_size=samples_per_bus)
        for idx_pair in sample_indexes:
            # автобус бүр хамгийн оптимал утгаар хөдлөнө гэж үзээд BUS_ID-г хаслаа
            # DAY_OF_WEEK, TIME, BUSROUTE_ID, BUSSTOP_ID, TIMESTAMP_DIFF
            dataset.append(bus_history[['DAY_OF_WEEK', 'TIME', 'BUSROUTE_ID', 'BUSSTOP_ID', 'TIMESTAMP_DIFF']].iloc[idx_pair[0]:idx_pair[1]].values.tolist())
    #break

print("done, total generated", len(dataset))
```

```python
np.save("./seq_dataset.npy", np.array(dataset))
```
