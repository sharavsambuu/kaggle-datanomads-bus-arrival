```python
import pandas as pd
import numpy as np
import time
import pickle
from sklearn.model_selection import train_test_split, GridSearchCV
from catboost import CatBoostRegressor
from sklearn.metrics import mean_absolute_error
```

```python
df = pd.read_csv("./out/cleaned_train.csv")
df
```

```python
inputs = df.drop(["TIMESTAMP_DIFF"], axis=1)
labels = df['TIMESTAMP_DIFF']
```

```python
#del model
model = CatBoostRegressor(
    iterations    = 100000,
    learning_rate = 0.7,
    depth         = 10,
    l2_leaf_reg   = 3,
    loss_function = 'MAE',
    task_type     = 'GPU'
  )
```

```python
model.fit(inputs, labels, verbose=100);
```

```python
model.save_model('catboost_10000')
```

```python
#load trained model
nmodel = CatBoostRegressor()
nmodel.load_model('catboost_10000')
```

```python
print(nmodel.predict(inputs.iloc[:5]))
```
