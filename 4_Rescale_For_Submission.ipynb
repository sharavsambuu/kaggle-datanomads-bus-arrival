```python
from datetime import datetime
import numpy as np
import pandas as pd
import random
```

```python
# Дараах kernel-үүдэд эцсийн data submit хийхэд шаардлагатай туршилтууд агуулагдана
```

```python
# Одоогийн хугацааг аваад timestamp-руу хөрвүүлж нэг үзэх
now_dt = datetime.now()
now_ts = datetime.timestamp(now_dt)

print(now_dt)
print(now_ts)
```

```python
# Санамсаргүйгээр хэдэн секунтууд сонгоод шинэ timestamp-уудын жагсаалтыг 
# өмнөх үүсгэсэн хугацааны timestamp дээр нэмэх байдлаар үүсгэх
ts_list = [now_ts]
ts      = now_ts
for i in range(0, 10):
    ts = ts + random.randrange(20, 40)
    ts_list.append(ts)

ts_list
```

```python
# Үүсгэсэн unix timestamp-уудын жагсаалтуудыг буцаагаад datetime
# объектүүдрүү хөрвүүлж нэг харах
for dt in [datetime.utcfromtimestamp(i) for i in ts_list]:
    print(dt)
```

```python
# Эдгээр timestamp-уудын жагсаалтыг тодорхой хугацааны завсарт
# шахаж хөрвүүлэх туршилт хийхийн тулд завсарыг илэрхийлэх
# timestamp-уудыг үүсгэх
scale_min = now_ts
scale_max = now_ts+8*60 
print(scale_min, scale_max)
```

```python
# Жагсаалтаа numpy массив болгож үүсгэх
ts_list_np = np.array(ts_list)
ts_list_np
```

```python
# timestamp-уудын жагсаалтыг scale_min болон scale_max утгуудын хооронд хүчээр align хийх
# буюу нэг ёсондоо дөхөлт хийлгэн агшаах/сунгах.
rescaled_ts_list = np.interp(ts_list_np, (ts_list_np.min(), ts_list_np.max()), (scale_min, scale_max))
```

```python
# Өмнөх timestamp-уудын жагсаалтыг datetime хөрвүүлгээр харах
print("===== before scale =======")
for dt in [datetime.utcfromtimestamp(i) for i in ts_list]:
    print(dt)

# Тодорхой завсарт агшаах/сунгасны дараах үр дүнг харах
print("===== after rescaled =====")
for dt in [datetime.utcfromtimestamp(i) for i in rescaled_ts_list]:
    print(dt)
    
```

```python
# Бодит test датаг дуурайлгахын тулд шинээр pandas dataframe үүсгэх
d  = {'TIMESTAMP': ts_list, 'DATETIME': [datetime.utcfromtimestamp(i) for i in ts_list]}
df = pd.DataFrame(data=d)
df
```

```python
# test датаг дуурайлган зарим мөрүүдийг нь сохлох
df.iloc[1  ] = np.nan
df.iloc[4:7] = np.nan
df.iloc[9  ] = np.nan
df
```

```python
# backup timestamp list-ийг эцсийн submission-д scale хийх туршилтын 
# хүрээнд багана байдлаар хадгалж авах. Энэ баганы scale хийж харна.
another_df = pd.DataFrame(
    data={
        'TIMESTAMP': rescaled_ts_list,
        'DATETIME' : [datetime.utcfromtimestamp(i) for i in rescaled_ts_list]
    })
df['TIMESTAMP_BK'] = another_df['TIMESTAMP']
df['DATETIME_BK' ] = another_df['DATETIME' ]
df
```

```python
# TIMESTAMP багана дээр суурилан сохолсон завсаруудыг илрүүлэн хадгалах
gap_start = None
gap_end   = None
gap_list  = []

for idx, row in df.iterrows():
    timestamp = row['TIMESTAMP']
    if (pd.isna(timestamp)):
        prev_timestamp = df.iloc[idx-1]['TIMESTAMP']
        if (not pd.isna(prev_timestamp)):
            gap_start = idx-1
            pass
        next_timestamp = df.iloc[idx+1]['TIMESTAMP']
        if (not pd.isna(next_timestamp)):
            gap_end = idx+1
            gap_list.append((gap_start, gap_end))
            
gap_list
```

```python
# Сохолсон завсарт байгаа эхлэл төгсөл timestamp-уудын дагуу
# backup timestamp-уудыг scale хийж хадгалах. 
# Ингээд энэ хэсэгт туршилт дуусч байна.
for gap in gap_list:
    gap_start = gap[0]
    gap_end   = gap[1]
    scale_min = df.iloc[gap_start]['TIMESTAMP']
    scale_max = df.iloc[gap_end  ]['TIMESTAMP']
    bk_timestamps = df.iloc[gap_start:gap_end+1]['TIMESTAMP_BK'].values
    
    rescaled = np.interp(
        bk_timestamps, 
        (bk_timestamps.min(), bk_timestamps.max()), 
        (scale_min, scale_max)
    )
    iter_count = gap_end - gap_start
    for i in range(0, iter_count+1):
        idx = gap_start+i
        df.at[idx, 'TIMESTAMP_SCALED'] = rescaled[i]
        df.at[idx, 'DATETIME_SCALED' ] = datetime.utcfromtimestamp(rescaled[i])
    
df
```

```python
# Дараах kernel-үүдээс эхлэн BusNet-ээр predict хийж бөглөсөн timestamp-уудыг 
# test дээрхи завсаруудын дагуу scale хийн эцсийн submission хийх үр дүн болгон бэлтгэнэ.
```

```python
# BusNet-ээр бөглөсөн timestamp-уудыг эцсийн submission-д бэлтгэх
test_df = pd.read_csv("./out/test_df_preadjustion.csv")
```

```python
test_df.head(15)
```

```python
# Завсруудын эхлэл төгсгөлүүдийг илрүүлж цуглуулах
gap_list = []
for idx, row in test_df.iterrows():
    timestamp = row['TIMESTAMP_BK']
    if (pd.isna(timestamp)):
        prev_timestamp = test_df.iloc[idx-1]['TIMESTAMP_BK']
        if (not pd.isna(prev_timestamp)):
            gap_start = idx-1
            pass
        next_timestamp = test_df.iloc[idx+1]['TIMESTAMP_BK']
        if (not pd.isna(next_timestamp)):
            gap_end = idx+1
            gap_list.append((gap_start, gap_end))
            
len(gap_list)
```

```python
gap_list[:5]
```

```python
# Test дэх завсаруудын дагуу predict хийсэн timestamp-уудыг дахин scale хийх
for gap in gap_list:
    gap_start     = gap[0]
    gap_end       = gap[1]
    scale_min     = test_df.iloc[gap_start]['TIMESTAMP_BK']
    scale_max     = test_df.iloc[gap_end  ]['TIMESTAMP_BK']
    bk_timestamps = test_df.iloc[gap_start:gap_end+1]['TIMESTAMP'].values
    
    bk_timestamps[0] = scale_min
    
    rescaled = np.interp(
        bk_timestamps, 
        (bk_timestamps.min(), bk_timestamps.max()), 
        (scale_min, scale_max)
    )
    iter_count = gap_end - gap_start
    for i in range(0, iter_count+1):
        idx = gap_start+i
        test_df.at[idx, 'TIMESTAMP_SCALED'] = rescaled[i]
        test_df.at[idx, 'DATETIME_SCALED' ] = datetime.utcfromtimestamp(rescaled[i])
        
```

```python
# Зарим тохиолдолд test дотор дараалсан өгөгдлүүд байгаа, энэ үед завсруудын дагуу
# scale хийж хадгалагдсан өгөгдлүүд TIMESTAMP_SCALED баганад хадгалагдаад бусад нь
# хадгалагдахгүй үлдэх боломжтой. Энэ тохиолдолд TIMESTAMP_BK-ээс утга авч хадгалах хэрэгтэй.
for idx, row in test_df.iterrows():
    timestamp_scaled = row['TIMESTAMP_SCALED']
    if (pd.isna(timestamp_scaled)):
        test_df.at[idx, 'TIMESTAMP_SCALED'] = row['TIMESTAMP_BK']
```

```python
# Одоо байгаа test_df-ээ юу ч болсон хадгалах
test_df.to_csv("./out/test_df_timestamp_adjusted.csv")

# Эцсийн submission-д явуулах датаг хадгалах. 
# TIMESTAMP_SCALED баганы нэрийг TIMESTAMP болгож өөрчлөх хэрэгтэй.
test_df[['index', 'TIMESTAMP_SCALED']].to_csv("./out/output_ready.csv", index=False)
```
